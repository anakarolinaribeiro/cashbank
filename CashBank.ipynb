```python
menu = """

[1]Depositar
[2]Sacar
[3]Extrato
[4]Sair


=> """

saldo=0
limite=500
extrato=" "
numero_saques=0
LIMITE_SAQUES=3

while True:
    
        opcao=input(menu)
    
        if opcao == "1":
            valor= float(input("Qual valor você deseja depositar? "))
        
        
            if valor >0:
                saldo+=valor
                extrato+= f"Depósito: R$ {valor:.2f}"
            
            else: 
                print("Erro em operação! valor informado é inválido.")
            
        elif opcao == "2":
            valor= float(input("Qual valor você deseja sacar? "))
        
            excedeu_saldo = valor > saldo
            excedeu_limite = valor > limite
            exedeu_saques = numero_saques>= LIMITE_SAQUES
        
            if  excedeu_saldo:
                print("Falha em operação! Não há saldo suficiente.")
            
            elif  excedeu_limite:
                print("Falha em operação! Valor de saque é excedente ao limite permitido.")
            
            elif excedeu_saques:
                print("Falha em operação! Número máximo de saques excedido")
            
            elif valor>0:
                saldo-=valor
                extrato+= f"Saque: R$ {valor:.2f}\n"
                numero_saques += 1
        
            else:
                print("Falha em operação! Valor informado inválido")
            
        elif opcao == "3":            
            print("\n _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ ")
            print("\n _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ ")
            print("\n _ _ _ _ _ _ _ _ _ _ EXTRATO _ _ _ _ _ _ _ _ _ _ ")
            print("Não foram realizadas movimentações. " if not extrato else extrato)
            print(f"\nSaldo: R$ {saldo:.2f}")
            print("\n _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ ")
            print("\n _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ ")
            print("\n _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ _ ")
    
           
        
        elif opcao == "4":
            break
        
        else:
            print("Operação inválida, selecione novamente a opção desejada")
    
```
